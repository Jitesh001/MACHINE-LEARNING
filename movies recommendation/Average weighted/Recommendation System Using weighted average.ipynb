# Basic Recommendation System

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

pd.pandas.set_option('display.max_columns', None)
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv("tmdb_5000_movies.csv")

print(credits.shape)
print(movies.shape)
```

```python
credits.head()
```

```python
movies.head()
```

```python
# In above two dataset id column is comman so we can merge both dataframe using id column
# rename the movie_id with id
credits_rename = credits.rename(index = str, columns = {"movie_id":"id"})
movies_df_merged = movies.merge(credits_rename, on = "id")
movies_df_merged.head()
```

```python
#from above datset we can see that columns like homepag, title_x, title_y, status, 
#production_countries not providing much information we are dropping them
movies_cleaned_df = movies_df_merged.drop(columns = ['homepage', 'title_x', 'title_y', 'status',
                                                     'production_countries'])
movies_cleaned_df.head()
```

```python
#lets check for null values
movies_cleaned_df.isnull().sum()
```

tagline contains 844 null values but its not imp feature for us now so ignore it
we are building recommendation on the basis of ratings and popularity, so focusing on these features

### Using Weighted average for each movie's Average Rating

```python
R = movies_cleaned_df["vote_average"]
v = movies_cleaned_df["vote_count"]
m = movies_cleaned_df['vote_count'].quantile(0.7) #minimum 70% of total votes
c = movies_cleaned_df["vote_average"].mean()
```

```python
movies_cleaned_df['weighted_average'] = ((R*v)+(c*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movies_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending = False)
movies_sorted_ranking[['original_title', 'vote_count', 'vote_average', 
                       'weighted_average', 'popularity']].head(10)
```

```python
weighted_avg = movies_cleaned_df.sort_values('weighted_average', ascending = False)
plt.figure(figsize = (10, 5))
axis1 = sns.barplot(x = weighted_avg['weighted_average'].head(10),
                    y = weighted_avg['original_title'].head(10), 
                    data = weighted_avg)
plt.xlim(4, 10)
plt.xlabel("Rating", weight = 'bold')
plt.ylabel("Title",  weight = 'bold')
plt.title("Best Movies by average votes",  weight = 'bold')
plt.savefig("best_movies.png",  weight = 'bold')
```

```python
# Lets plot on the basis of popularity
popular_df = movies_cleaned_df.sort_values('popularity', ascending = False)
plt.figure(figsize=(10, 5))
axis1 = sns.barplot(x = popular_df['popularity'].head(10), y = popular_df['original_title'].head(10),
                   data = popular_df)
plt.xlabel('popularity', weight = 'bold')
plt.ylabel('title', weight = 'bold')
plt.title('movies popularity', weight = 'bold')
plt.savefig('popularMovies.png')
```

### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average', 'popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df, columns = ['weighted_average', 'popularity'] )
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average']*0.5 +\
                             movies_cleaned_df['normalized_popularity']*0.5
    
movies_scored_df = movies_cleaned_df.sort_values('score', ascending = False)
movies_scored_df[['original_title', 'normalized_weight_average', 
                  'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
